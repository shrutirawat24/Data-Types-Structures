Q1 1. Explain the different types of data (qualitative and quantitative) and provide examples of each. Discuss
nominal, ordinal, interval, and ratio scales.

Ans Data can be classified into two main types: Qualitative and Quantitative.

1. Qualitative Data (Categorical Data)
Qualitative data describes characteristics or attributes that cannot be measured numerically
Eg. Colors of car: Red, Blue, Green ;
Gender: Male, Female, Non-binary


2. Quantitative Data (Numerical Data)
Quantitative data consists of numerical values that can be counted or measured.
Eg. Number of cars in a parking lot

The scales of measurement:
1. Nominal Scale
There is no inherent order or numerical value associated with the categories. The numbers assigned are merely labels or codes and cannot be used for mathematical operations.

2. Ordinal Scale
The ordinal scale allows for categorization and ranking of data, meaning there's a meaningful order among the categories.

3. Interval Scale
The interval scale possesses all the characteristics of nominal and ordinal scales, plus it has equal intervals between values.

4. Ratio Scale
 It includes all the properties of the nominal, ordinal, and interval scales, and crucially, it has a "true zero" point. This allows for meaningful ratios to be calculated.

Q3. Explain the concept of dispersion. How do variance and standard deviation measure the spread of data?

Ans Dispersion refers to the extent to which data points in a distribution are stretched out or squeezed together. While measures of central tendency (like the mean, median, and mode) tell us about the "typical" value, measures of dispersion tell us how much the individual data points deviate from that central value
Both variance and standard deviation quantify the average amount by which individual data points deviate from the mean of the dataset.

Q7. What is the interquartile range (IQR), and how is it used to detect outliers?

Ans The Interquartile Range (IQR) is a measure of statistical dispersion, specifically representing the spread of the middle 50% of a dataset. It is not affected by extreme values or outliers, unlike the full range (maximum - minimum).
The formula for the IQR is simply the difference between the third quartile (Q3) and the first quartile (Q1):IQR=Q3−Q1


The IQR is commonly used to define a range within which most of the data points are expected to lie. Any data point that falls significantly outside this range is considered a potential outlier. This method is particularly useful because it's based on the middle 50% of the data, making it less sensitive to extreme values than methods based on the mean and standard deviation.

Q9. Explain the properties of the normal distribution and the empirical rule (68-95-99.7 rule).

Ans The properties are:

 1. Bell-Shaped and Symmetrical: When plotted on a graph, the normal distribution forms a characteristic bell-shaped curve. It is perfectly symmetrical around its center, meaning that if you fold the curve at the mean, the left and right sides would be mirror images of each other.


 2. Mean, Median, and Mode are Equal: Due to its perfect symmetry, the mean, median, and mode of a normal distribution are all located at the exact same point – the center of the distribution, which is also its peak.

3. Asymptotic to the X-axis: The tails of the normal distribution curve extend infinitely in both directions, getting closer and closer to the x-axis but never actually touching it. This implies that theoretically, any value, no matter how extreme, has a non-zero probability of occurring
4. Unimodal: It has only one peak, making it a unimodal distribution.

The Empirical Rule, also known as the 68-95-99.7 rule or the "Three Sigma Rule," is a useful guideline that describes the percentage of data that falls within specific numbers of standard deviations from the mean in a normal distribution. It's an approximation, but a very good one for understanding the spread of normally distributed data.

Q11. Explain what a random variable is and differentiate between discrete and continuous random variables.


Ans A random variable is a numerical description of the outcome of a random phenomenon or experiment. It's not a variable in the algebraic sense (where 'x' represents a single unknown value), but rather a function that assigns a numerical value to each possible outcome in a sample space.

Difference:

1. Discrete Random Variables
A discrete random variable is a random variable that can take on a finite number of values or a countably infinite number of values. This means the values are distinct and separate, often integers, and there are "gaps" between consecutive possible values.

2. Continuous Random Variables
A continuous random variable is a random variable that can take on any value within a given range or interval. These values are not countable; they can be measured.

Q12. Provide an example dataset, calculate both covariance and correlation, and interpret the results.

Ans


```python
import numpy as np

# Data
temperature = np.array([20, 22, 25, 23, 28]) # X (independent variable)
ice_cream_sales = np.array([3, 4, 5, 4.5, 6]) # Y (dependent variable)

# --- 1. Calculate Covariance ---
# numpy.cov returns a covariance matrix. For two 1D arrays, the off-diagonal
# element [0, 1] or [1, 0] is the covariance between the two arrays.
# It uses n-1 in the denominator by default (sample covariance).
covariance_matrix = np.cov(temperature, ice_cream_sales)
covariance = covariance_matrix[0, 1]

print(f"Dataset:")
print(f"  Temperature (X): {temperature}")
print(f"  Ice Cream Sales (Y): {ice_cream_sales}")
print("-" * 30)
print(f"1. Covariance:")
print(f"   Calculated Covariance (s_xy): {covariance:.3f}")

# --- Interpretation of Covariance ---
if covariance > 0:
    cov_interpretation = "Positive covariance: As temperature increases, ice cream sales tend to increase."
elif covariance < 0:
    cov_interpretation = "Negative covariance: As temperature increases, ice cream sales tend to decrease."
else:
    cov_interpretation = "Covariance is zero: No linear relationship between temperature and ice cream sales."
print(f"   Interpretation: {cov_interpretation}")

print("-" * 30)

# --- 2. Calculate Correlation ---
# numpy.corrcoef returns a correlation coefficient matrix. For two 1D arrays,
# the off-diagonal element [0, 1] or [1, 0] is the Pearson correlation coefficient.
correlation_matrix = np.corrcoef(temperature, ice_cream_sales)
correlation = correlation_matrix[0, 1]

print(f"2. Correlation (Pearson Product-Moment Correlation Coefficient):")
print(f"   Calculated Correlation (r_xy): {correlation:.3f}")

# --- Interpretation of Correlation ---
print(f"   Interpretation:")
if correlation > 0:
    print(f"     Direction: Positive linear relationship. As temperature increases, ice cream sales tend to increase.")
else:
    print(f"     Direction: Negative linear relationship. As temperature increases, ice cream sales tend to decrease.")

abs_correlation = abs(correlation)
if abs_correlation >= 0.9:
    strength = "Very strong"
elif abs_correlation >= 0.7:
    strength = "Strong"
elif abs_correlation >= 0.5:
    strength = "Moderate"
elif abs_correlation >= 0.3:
    strength = "Weak"
else:
    strength = "Very weak or no"

print(f"     Strength: {strength} linear relationship.")
if correlation > 0.95: # Specific for very high values
    print(f"     This means temperature is an extremely strong predictor of ice cream sales for this shop.")
```
