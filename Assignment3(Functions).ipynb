Q1  Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list.

Ans

```python
def sum_of_even_numbers(numbers):
    # Initialize the sum variable
    sum_even = 0

    # Loop through the list and add even numbers to the sum
    for num in numbers:
        if num % 2 == 0:
            sum_even += num

    return sum_even

# Example
numbers = [47, 11, 42, 13, 8]
result = sum_of_even_numbers(numbers)
print("Sum of even numbers:", result)
```

Q4  Write a Python function that checks if a given number is prime or not from 1 to 200.

Ans

```python
def is_prime(number):
    if number <= 1:
        return False
    for i in range(2, int(number ** 0.5) + 1):
        if number % i == 0:
            return False
    return True

# Check for prime numbers from 1 to 200
for num in range(1, 201):
    if is_prime(num):
        print(f"{num} is prime")
```

Q5 Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.

Ans


```python
class FibonacciIterator:
    def __init__(self, terms):
        self.terms = terms
        self.current = 0
        self.a, self.b = 0, 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.current < self.terms:
            fibonacci_number = self.a
            self.a, self.b = self.b, self.a + self.b
            self.current += 1
            return fibonacci_number
        else:
            raise StopIteration

# Example
fibonacci_sequence = FibonacciIterator(10)
for num in fibonacci_sequence:
    print(num)
```

Q6  Write a generator function in Python that yields the powers of 2 up to a given exponent.

Ans

```python
def powers_of_2(exponent):
    for i in range(exponent + 1):
        yield 2 ** i

# Example
for power in powers_of_2(5):
    print(power)
```

Q7  Implement a generator function that reads a file line by line and yields each line as a string.

Ans


```python
def read_file_line_by_line(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line.strip()

# Example
file_path = 'example.txt'  #your file path here
for line in read_file_line_by_line(file_path):
    print(line)
```

Q9 Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

Ans

```python
# List of temperatures in Celsius
celsius_temperatures = [0, 20, 30, 40, 100]

# Function to convert Celsius to Fahrenheit
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

# Using map() to apply the conversion function to each element in the list
fahrenheit_temperatures = list(map(celsius_to_fahrenheit, celsius_temperatures))

# Print the converted temperatures
print(fahrenheit_temperatures)
```

Q10 Create a Python program that uses `filter()` to remove all the vowels from a given string.

Ans

```python
# Function to check if a character is not a vowel
def is_not_vowel(char):
    vowels = "aeiouAEIOU"
    return char not in vowels

# Given string
input_string = "Hello, World!"

# Using filter() to remove vowels
filtered_string = ''.join(filter(is_not_vowel, input_string))

# Print the result
print(filtered_string)
```
