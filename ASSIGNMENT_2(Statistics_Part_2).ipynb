Q1Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and
interpret the results


```python
import numpy as np
from scipy import stats

def perform_z_test(sample_mean, population_mean, population_std_dev, sample_size, alpha=0.05, two_tailed=True):
    """
    Performs a Z-test to compare a sample mean to a known population mean.

    Args:
        sample_mean (float): The mean of the sample.
        population_mean (float): The known mean of the population.
        population_std_dev (float): The known standard deviation of the population.
        sample_size (int): The number of observations in the sample.
        alpha (float): The significance level (default is 0.05).
        two_tailed (bool): If True, performs a two-tailed test; otherwise, a one-tailed test.

    Returns:
        dict: A dictionary containing the Z-statistic, p-value, and the test decision.
    """
    # Calculate the standard error of the mean
    # Standard Error (SE) = Population Standard Deviation / sqrt(Sample Size)
    standard_error = population_std_dev / np.sqrt(sample_size)

    # Calculate the Z-statistic
    # Z = (Sample Mean - Population Mean) / Standard Error
    z_statistic = (sample_mean - population_mean) / standard_error

    # Calculate the p-value
    if two_tailed:
        # For a two-tailed test, we look at the probability in both tails.
        # stats.norm.sf(abs(z_statistic)) gives the survival function (1 - CDF) for the positive Z-value,
        # which is the area in the right tail. We multiply by 2 for both tails.
        p_value = 2 * stats.norm.sf(abs(z_statistic))
    else:
        # For a one-tailed test, we check the direction of the alternative hypothesis.
        # If sample_mean > population_mean, it's a right-tailed test.
        # If sample_mean < population_mean, it's a left-tailed test.
        if z_statistic > 0:
            # Right-tailed test: p-value is the area in the right tail
            p_value = stats.norm.sf(z_statistic)
        else:
            # Left-tailed test: p-value is the area in the left tail
            p_value = stats.norm.cdf(z_statistic)

    # Make a decision based on the p-value and alpha
    decision = "Fail to reject the null hypothesis"
    if p_value < alpha:
        decision = "Reject the null hypothesis"

    return {
        "z_statistic": z_statistic,
        "p_value": p_value,
        "alpha": alpha,
        "decision": decision,
        "test_type": "Two-tailed" if two_tailed else "One-tailed"
    }

# --- Example Usage ---

# Scenario 1: Two-tailed test (default)
# A factory produces light bulbs with a known average lifespan of 1500 hours and a standard deviation of 100 hours.
# A new batch of 50 bulbs is tested, and their average lifespan is 1530 hours.
# Is there a significant difference in the average lifespan of this new batch compared to the known population?
print("--- Scenario 1: Two-tailed Z-test ---")
sample_mean_1 = 1530
population_mean_1 = 1500
population_std_dev_1 = 100
sample_size_1 = 50
alpha_1 = 0.05

results_1 = perform_z_test(sample_mean_1, population_mean_1, population_std_dev_1, sample_size_1, alpha=alpha_1)
print(f"Sample Mean: {sample_mean_1}")
print(f"Population Mean: {population_mean_1}")
print(f"Population Standard Deviation: {population_std_dev_1}")
print(f"Sample Size: {sample_size_1}")
print(f"Significance Level (alpha): {results_1['alpha']}")
print(f"Test Type: {results_1['test_type']}")
print(f"Z-statistic: {results_1['z_statistic']:.4f}")
print(f"P-value: {results_1['p_value']:.4f}")
print(f"Decision: {results_1['decision']}")
print("\n" + "="*50 + "\n")

# Scenario 2: One-tailed test (right-tailed)
# A company claims its new fertilizer increases plant height.
# The known average height of plants without fertilizer is 20 cm with a standard deviation of 3 cm.
# A sample of 30 plants treated with the new fertilizer has an average height of 21.5 cm.
# Does the fertilizer significantly increase plant height?
print("--- Scenario 2: One-tailed (Right-tailed) Z-test ---")
sample_mean_2 = 21.5
population_mean_2 = 20
population_std_dev_2 = 3
sample_size_2 = 30
alpha_2 = 0.01 # Stricter alpha
two_tailed_2 = False # One-tailed test

results_2 = perform_z_test(sample_mean_2, population_mean_2, population_std_dev_2, sample_size_2, alpha=alpha_2, two_tailed=two_tailed_2)
print(f"Sample Mean: {sample_mean_2}")
print(f"Population Mean: {population_mean_2}")
print(f"Population Standard Deviation: {population_std_dev_2}")
print(f"Sample Size: {sample_size_2}")
print(f"Significance Level (alpha): {results_2['alpha']}")
print(f"Test Type: {results_2['test_type']}")
print(f"Z-statistic: {results_2['z_statistic']:.4f}")
print(f"P-value: {results_2['p_value']:.4f}")
print(f"Decision: {results_2['decision']}")
print("\n" + "="*50 + "\n")

# Scenario 3: One-tailed test (left-tailed)
# A new manufacturing process is supposed to reduce defects.
# The old process had an average of 5 defects per batch with a standard deviation of 1.5 defects.
# A sample of 40 batches from the new process has an average of 4.2 defects.
# Does the new process significantly reduce defects?
print("--- Scenario 3: One-tailed (Left-tailed) Z-test ---")
sample_mean_3 = 4.2
population_mean_3 = 5
population_std_dev_3 = 1.5
sample_size_3 = 40
alpha_3 = 0.05
two_tailed_3 = False # One-tailed test

results_3 = perform_z_test(sample_mean_3, population_mean_3, population_std_dev_3, sample_size_3, alpha=alpha_3, two_tailed=two_tailed_3)
print(f"Sample Mean: {sample_mean_3}")
print(f"Population Mean: {population_mean_3}")
print(f"Population Standard Deviation: {population_std_dev_3}")
print(f"Sample Size: {sample_size_3}")
print(f"Significance Level (alpha): {results_3['alpha']}")
print(f"Test Type: {results_3['test_type']}")
print(f"Z-statistic: {results_3['z_statistic']:.4f}")
print(f"P-value: {results_3['p_value']:.4f}")
print(f"Decision: {results_3['decision']}")
print("\n" + "="*50 + "\n")
```

Q2 Simulate random data to perform hypothesis testing and calculate the corresponding P-value using Python@

```python
import numpy as np
from scipy import stats

# Set random seed for reproducibility
np.random.seed(42)

# Simulate two random samples
# Sample A (e.g., control group)
sample_a = np.random.normal(loc=50, scale=10, size=30)

# Sample B (e.g., treatment group)
sample_b = np.random.normal(loc=55, scale=10, size=30)

# Perform two-sample t-test
t_stat, p_value = stats.ttest_ind(sample_a, sample_b)

# Output results
print("Sample A mean:", np.mean(sample_a))
print("Sample B mean:", np.mean(sample_b))
print("T-statistic:", t_stat)
print("P-value:", p_value)

# Interpret the P-value
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: Significant difference between groups.")
else:
    print("Fail to reject the null hypothesis: No significant difference between groups.")
```

Q4  Perform a two-tailed Z-test using Python and visualize the decision region on a plot.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters
population_mean = 100
population_std = 15
alpha = 0.05  # Significance level (two-tailed)
sample = np.array([102, 98, 105, 95, 100, 110, 108, 99, 101, 107])

# Sample statistics
sample_mean = np.mean(sample)
sample_size = len(sample)

# Z-statistic
z = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# Two-tailed P-value
p_value = 2 * (1 - norm.cdf(abs(z)))

# Critical Z values
z_critical = norm.ppf(1 - alpha / 2)

# Output results
print(f"Sample Mean = {sample_mean:.2f}")
print(f"Z-statistic = {z:.2f}")
print(f"P-value = {p_value:.4f}")
print(f"Z-critical (±): ±{z_critical:.2f}")

# Decision
if abs(z) > z_critical:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")

# === Visualization ===
# Plot the standard normal distribution
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Standard Normal Distribution')

# Fill rejection regions
plt.fill_between(x, y, where=(x < -z_critical), color='red', alpha=0.5, label='Rejection Region (left)')
plt.fill_between(x, y, where=(x > z_critical), color='red', alpha=0.5, label='Rejection Region (right)')

# Draw Z-statistic
plt.axvline(z, color='blue', linestyle='--', label=f'Z-statistic = {z:.2f}')

# Add annotations
plt.axvline(-z_critical, color='black', linestyle=':', label=f'Z-critical = ±{z_critical:.2f}')
plt.axvline(z_critical, color='black', linestyle=':')

plt.title('Two-tailed Z-Test and Decision Regions')
plt.xlabel('Z-value')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

Q5 Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.


```python


def plot_type_errors(mu_null, mu_alt, sigma, n, alpha, tail='two'):
  """
  Calculates and visualizes Type 1 and Type 2 errors for a hypothesis test.

  Args:
    mu_null (float): The population mean under the null hypothesis (H0).
    mu_alt (float): The true population mean under the alternative hypothesis (H1).
    sigma (float): The known population standard deviation.
    n (int): The sample size.
    alpha (float): The significance level (Type 1 error rate).
    tail (str): The type of test: 'two', 'right', or 'left'.
  """
  se = sigma / np.sqrt(n)

  # Define the standard error of the mean
  x_null = np.linspace(mu_null - 4 * se, mu_null + 4 * se, 1000)
  y_null = norm.pdf(x_null, mu_null, se)

  x_alt = np.linspace(mu_alt - 4 * se, mu_alt + 4 * se, 1000)
  y_alt = norm.pdf(x_alt, mu_alt, se)

  plt.figure(figsize=(12, 7))
  plt.plot(x_null, y_null, label=f'H0: Mean = {mu_null}', color='blue')
  plt.plot(x_alt, y_alt, label=f'H1: Mean = {mu_alt} (True Mean)', color='green')

  # Find critical value(s) based on alpha and tail
  if tail == 'two':
    z_crit_lower = norm.ppf(alpha / 2, mu_null, se)
    z_crit_upper = norm.ppf(1 - alpha / 2, mu_null, se)
    critical_values = [z_crit_lower, z_crit_upper]
    rejection_region_h0 = (x_null < z_crit_lower) | (x_null > z_crit_upper)
    non_rejection_region_h0 = (x_null >= z_crit_lower) & (x_null <= z_crit_upper)
    rejection_region_h1 = (x_alt < z_crit_lower) | (x_alt > z_crit_upper)
    non_rejection_region_h1 = (x_alt >= z_crit_lower) & (x_alt <= z_crit_upper)

    # Visualize critical values
    plt.axvline(z_crit_lower, color='red', linestyle='--', label=f'Critical Value (Lower): {z_crit_lower:.2f}')
    plt.axvline(z_crit_upper, color='red', linestyle='--', label=f'Critical Value (Upper): {z_crit_upper:.2f}')

    # Calculate Type II error (Beta)
    # Probability of failing to reject H0 when H1 is true
    beta = norm.cdf(z_crit_upper, mu_alt, se) - norm.cdf(z_crit_lower, mu_alt, se)

  elif tail == 'right':
    z_crit = norm.ppf(1 - alpha, mu_null, se)
    critical_values = [z_crit]
    rejection_region_h0 = (x_null > z_crit)
    non_rejection_region_h0 = (x_null <= z_crit)
    rejection_region_h1 = (x_alt > z_crit)
    non_rejection_region_h1 = (x_alt <= z_crit)

    # Visualize critical value
    plt.axvline(z_crit, color='red', linestyle='--', label=f'Critical Value: {z_crit:.2f}')

    # Calculate Type II error (Beta)
    # Probability of failing to reject H0 (sample mean <= z_crit) when H1 is true
    beta = norm.cdf(z_crit, mu_alt, se)

  elif tail == 'left':
    z_crit = norm.ppf(alpha, mu_null, se)
    critical_values = [z_crit]
    rejection_region_h0 = (x_null < z_crit)
    non_rejection_region_h0 = (x_null >= z_crit)
    rejection_region_h1 = (x_alt < z_crit)
    non_rejection_region_h1 = (x_alt >= z_crit)

    # Visualize critical value
    plt.axvline(z_crit, color='red', linestyle='--', label=f'Critical Value: {z_crit:.2f}')

    # Calculate Type II error (Beta)
    # Probability of failing to reject H0 (sample mean >= z_crit) when H1 is true
    beta = 1 - norm.cdf(z_crit, mu_alt, se)

  else:
    raise ValueError("Invalid 'tail' argument. Use 'two', 'right', or 'left'.")

  # Visualize Type I Error (Alpha)
  plt.fill_between(x_null[rejection_region_h0], y_null[rejection_region_h0], color='red', alpha=0.3, label=f'Type I Error (Alpha) = {alpha:.4f}')

  # Visualize Type II Error (Beta)
  plt.fill_between(x_alt[non_rejection_region_h1], y_alt[non_rejection_region_h1], color='orange', alpha=0.5, label=f'Type II Error (Beta) = {beta:.4f}')

  plt.title(f'Type I and Type II Errors Visualization ({tail.capitalize()}-tailed Test)')
  plt.xlabel('Sample Mean')
  plt.ylabel('Probability Density')
  plt.legend()
  plt.grid(True)
  plt.show()

# --- Example Usage ---

# Scenario 1: Two-tailed test
print("--- Scenario 1: Two-tailed test for Type I and Type II Errors ---")
plot_type_errors(mu_null=100, mu_alt=105, sigma=15, n=30, alpha=0.05, tail='two')

# Scenario 2: Right-tailed test
print("\n--- Scenario 2: Right-tailed test for Type I and Type II Errors ---")
plot_type_errors(mu_null=50, mu_alt=53, sigma=10, n=25, alpha=0.01, tail='right')

# Scenario 3: Left-tailed test
print("\n--- Scenario 3: Left-tailed test for Type I and Type II Errors ---")
plot_type_errors(mu_null=80, mu_alt=77, sigma=8, n=40, alpha=0.10, tail='left')
```

Q9 Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

```python
import numpy as np
from scipy.stats import t

def confidence_interval(data, confidence=0.95):
    """
    Calculate the confidence interval for a sample mean using the t-distribution.

    Parameters:
    - data: list or numpy array of sample observations
    - confidence: confidence level (default is 0.95)

    Returns:
    - (lower_bound, upper_bound): tuple representing the confidence interval
    """
    sample = np.array(data)
    n = len(sample)
    mean = np.mean(sample)
    se = np.std(sample, ddof=1) / np.sqrt(n)
    t_crit = t.ppf((1 + confidence) / 2, df=n-1)
    margin = t_crit * se
    return mean - margin, mean + margin

# === Example Usage ===
np.random.seed(1)
sample_data = np.random.normal(loc=50, scale=10, size=25)
ci = confidence_interval(sample_data)
print(f"95% Confidence Interval for the mean: {ci}")
```

```python
import numpy as np
from scipy.stats import t

def margin_of_error(data, confidence=0.95):
    """
    Calculate the margin of error for the sample mean at a given confidence level.

    Parameters:
    - data: list or numpy array of sample observations
    - confidence: confidence level (default is 0.95)

    Returns:
    - margin of error
    """
    sample = np.array(data)
    n = len(sample)
    se = np.std(sample, ddof=1) / np.sqrt(n)  # standard error
    t_crit = t.ppf((1 + confidence) / 2, df=n - 1)
    moe = t_crit * se
    return moe

# === Example Usage ===
np.random.seed(10)
sample_data = np.random.normal(loc=70, scale=8, size=40)  # sample data
confidence_level = 0.95

moe = margin_of_error(sample_data, confidence=confidence_level)
print(f"Margin of Error at {confidence_level*100:.0f}% confidence: {moe:.3f}")
```

Q11  Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.

```python
def bayes_theorem(prior, sensitivity, false_positive_rate):
    """
    Apply Bayes' Theorem to calculate posterior probability.

    Parameters:
    - prior: P(D), prior probability of disease
    - sensitivity: P(Pos | D), probability of positive test given disease
    - false_positive_rate: P(Pos | No D), probability of positive test given no disease

    Returns:
    - posterior: P(D | Pos), probability of disease given positive test
    """
    p_not_disease = 1 - prior

    # Total probability of positive test
    p_positive = sensitivity * prior + false_positive_rate * p_not_disease

    # Bayes' Theorem
    posterior = (sensitivity * prior) / p_positive

    return posterior

# Example values
prior = 0.01               # 1% prevalence
sensitivity = 0.99         # 99% sensitivity
false_positive_rate = 0.05 # 5% false positive rate

posterior = bayes_theorem(prior, sensitivity, false_positive_rate)
print(f"Probability of having the disease after a positive test: {posterior:.4f} ({posterior*100:.2f}%)")
```

Q16  Implement an F-test using Python to compare the variances of two random samples.

```python
import numpy as np
from scipy.stats import f

# Step 1: Generate two random samples
np.random.seed(42)
sample1 = np.random.normal(loc=50, scale=5, size=30)
sample2 = np.random.normal(loc=52, scale=8, size=30)

# Step 2: Calculate sample variances
var1 = np.var(sample1, ddof=1)
var2 = np.var(sample2, ddof=1)

# Step 3: Calculate F-statistic (larger variance / smaller variance)
if var1 > var2:
    f_stat = var1 / var2
    dfn = len(sample1) - 1  # numerator degrees of freedom
    dfd = len(sample2) - 1  # denominator degrees of freedom
else:
    f_stat = var2 / var1
    dfn = len(sample2) - 1
    dfd = len(sample1) - 1

# Step 4: Calculate two-tailed p-value
p_value = 2 * min(f.cdf(f_stat, dfn, dfd), 1 - f.cdf(f_stat, dfn, dfd))

# Step 5: Output results
print("=== F-test for Equality of Variances ===")
print(f"Sample 1 Variance: {var1:.3f}")
print(f"Sample 2 Variance: {var2:.3f}")
print(f"F-statistic: {f_stat:.3f}")
print(f"Degrees of Freedom: {dfn}, {dfd}")
print(f"P-value: {p_value:.4f}")

# Step 6: Interpret results
alpha = 0.05
if p_value < alpha:
    print(f"\nReject H₀ (p < {alpha}): Variances are significantly different.")
else:
    print(f"\nFail to reject H₀ (p ≥ {alpha}): No significant difference in variances.")
```

Q19 Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA


```python
import numpy as np
import pandas as pd
from scipy.stats import shapiro, levene

def check_anova_assumptions(*groups):
    """
    Checks ANOVA assumptions:
    - Normality (Shapiro-Wilk test)
    - Equal variance (Levene's test)

    Args:
        *groups: Variable number of group arrays/lists

    Prints the test results and interpretation.
    """
    print("=== Checking ANOVA Assumptions ===\n")

    # Normality check for each group
    for i, group in enumerate(groups, 1):
        stat, p = shapiro(group)
        print(f"Group {i} Normality (Shapiro-Wilk): W={stat:.3f}, p={p:.4f}")
        if p > 0.05:
            print(f"  ✔ Group {i} looks normally distributed (fail to reject H0)")
        else:
            print(f"  ✖ Group {i} may not be normally distributed (reject H0)")

    print("\nChecking equal variance across groups (Levene's test)...")
    stat, p = levene(*groups)
    print(f"Levene's test: W={stat:.3f}, p={p:.4f}")
    if p > 0.05:
        print("  ✔ Variances are equal (fail to reject H0)")
    else:
        print("  ✖ Variances may be unequal (reject H0)")

    print("\nNote on independence:")
    print("  ✔ Independence is assumed from the study design (not tested from data).")
    print("    Ensure that samples are collected independently!")

# Example usage
if __name__ == "__main__":
    np.random.seed(42)
    g1 = np.random.normal(50, 5, 30)
    g2 = np.random.normal(52, 5, 30)
    g3 = np.random.normal(55, 5, 30)

    check_anova_assumptions(g1, g2, g3)
```

Q24  Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results

```python
import numpy as np
from scipy.stats import chi2

# Sample data (simulate or use real)
np.random.seed(42)
sample = np.random.normal(loc=100, scale=15, size=40)  # mean=100, std=15

# Known variance under null hypothesis
sigma0_squared = 225  # For example, hypothesized variance = 15^2

# Sample variance
s_squared = np.var(sample, ddof=1)

# Sample size
n = len(sample)

# Chi-square test statistic
chi_square_stat = (n - 1) * s_squared / sigma0_squared

# Degrees of freedom
df = n - 1

# Two-tailed p-value
p_value = 2 * min(chi2.cdf(chi_square_stat, df), 1 - chi2.cdf(chi_square_stat, df))

print("=== Variance Hypothesis Test (Chi-square) ===")
print(f"Sample variance: {s_squared:.3f}")
print(f"Chi-square statistic: {chi_square_stat:.3f}")
print(f"Degrees of freedom: {df}")
print(f"P-value: {p_value:.4f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject H₀: The population variance is significantly different from the hypothesized variance.")
else:
    print("Fail to reject H₀: Not enough evidence to say the population variance differs from the hypothesized variance.")
```

Q25 Write a Python script to perform a Z-test for comparing proportions between two datasets or groups.

```python
import numpy as np
from statsmodels.stats.proportion import proportions_ztest

def two_proportion_ztest(count1, nobs1, count2, nobs2, alternative='two-sided'):
    """
    Performs a two-proportion Z-test.

    Parameters:
        count1 (int): Number of successes in group 1
        nobs1 (int): Total observations in group 1
        count2 (int): Number of successes in group 2
        nobs2 (int): Total observations in group 2
        alternative (str): 'two-sided', 'smaller', or 'larger'

    Returns:
        z_stat (float): Z statistic
        p_value (float): P-value of the test
    """
    counts = np.array([count1, count2])
    nobs = np.array([nobs1, nobs2])

    z_stat, p_value = proportions_ztest(counts, nobs, alternative=alternative)
    return z_stat, p_value

# Example usage
if __name__ == "__main__":
    # Group 1: 45 successes out of 100
    count1 = 45
    nobs1 = 100

    # Group 2: 30 successes out of 90
    count2 = 30
    nobs2 = 90

    z_stat, p_value = two_proportion_ztest(count1, nobs1, count2, nobs2)
    print(f"Z-statistic: {z_stat:.3f}")
    print(f"P-value: {p_value:.4f}")

    alpha = 0.05
    if p_value < alpha:
        print("Reject the null hypothesis: proportions are significantly different.")
    else:
        print("Fail to reject the null hypothesis: no significant difference in proportions.")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f

# Simulate two samples
np.random.seed(42)
sample1 = np.random.normal(loc=0, scale=5, size=40)  # variance ~ 25
sample2 = np.random.normal(loc=0, scale=3, size=30)  # variance ~ 9

# Calculate sample variances
var1 = np.var(sample1, ddof=1)
var2 = np.var(sample2, ddof=1)

# Degrees of freedom
df1 = len(sample1) - 1
df2 = len(sample2) - 1

# Calculate F statistic (larger variance / smaller variance)
if var1 > var2:
    F = var1 / var2
    dfn, dfd = df1, df2
else:
    F = var2 / var1
    dfn, dfd = df2, df1

# Calculate p-value (two-tailed)
p_value = 2 * min(f.cdf(F, dfn, dfd), 1 - f.cdf(F, dfn, dfd))

print("=== F-test for Equality of Variances ===")
print(f"Sample variance 1: {var1:.3f}")
print(f"Sample variance 2: {var2:.3f}")
print(f"F statistic: {F:.3f}")
print(f"Degrees of freedom numerator: {dfn}")
print(f"Degrees of freedom denominator: {dfd}")
print(f"P-value (two-tailed): {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Reject null hypothesis: variances are significantly different.")
else:
    print("Fail to reject null hypothesis: no significant difference in variances.")

# Visualization of F-distribution and test statistic
x = np.linspace(0, 5, 500)
y = f.pdf(x, dfn, dfd)

plt.figure(figsize=(8, 5))
plt.plot(x, y, 'b-', label='F-distribution PDF')
plt.axvline(F, color='red', linestyle='--', label=f'F statistic = {F:.3f}')
plt.fill_between(x, 0, y, where=(x >= F), color='red', alpha=0.3)
plt.title('F-distribution with Test Statistic')
plt.xlabel('F value')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

Q27  Perform a Chi-square test for goodness of fit with simulated data and analyze the results

```python
import numpy as np
from scipy.stats import chisquare

# Step 1: Simulate observed data (counts in categories)
# Suppose we have 5 categories with these observed frequencies:
observed = np.array([18, 22, 20, 15, 25])

# Step 2: Define expected frequencies (under null hypothesis)
# For example, assume equal probability for each category
expected = np.array([20, 20, 20, 20, 20])

# Step 3: Perform Chi-square goodness-of-fit test
chi_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

print("=== Chi-square Goodness-of-Fit Test ===")
print(f"Chi-square statistic: {chi_stat:.3f}")
print(f"P-value: {p_value:.4f}")

# Step 4: Interpret results
alpha = 0.05
if p_value < alpha:
    print("Reject null hypothesis: observed distribution differs significantly from expected.")
else:
    print("Fail to reject null hypothesis: observed data fits expected distribution.")
```
